### Равномерная географическая сетка

Очевидно, что мы не можем придерживаться параллелей и меридианов для построения сетки, в противном случае размеры ячеек в метрах на экваторе будут отличаться от полюсов.

Среди возможных вариантов - использовать метод HEALPix, один из способов разбиения сферы на однородные области. Например, он использовался в одном из проектов NASA для отображения данных о реликтовом излучении, полученных с космического аппарата WMAP.

http://map.gsfc.nasa.gov/mission/observatory_pix.html

Ниже пример использования библиотеки healpy для Python. В частности, библиотека содержит методы по определению ячейки на сфере по угловым координатам произвольной точки и наоборот. С помощью параметра Nside можно менять количество ячеек, на которые разбивается сфера. 

http://healpy.readthedocs.org/en/latest/index.html - ссылка на Python API


```python
%matplotlib inline
import numpy as np
import healpy as hp
```

```python
NSIDE = 4
m = np.arange(hp.nside2npix(NSIDE))
hp.mollview(m, xsize=2000, title="Small number of cells")
```

```python
NSIDE = 16
m = np.arange(hp.nside2npix(NSIDE))
hp.mollview(m, xsize=2000, title="Medium cells number")
```
